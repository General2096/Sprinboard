When reading through any nutrition label, the most sought out piece of information is Calories. When choosing to reduce intake, Calories play such an important role to reach health goals. This dataset includes the nutritional information for Starbucks’ drink menu items. All nutritional information for drinks are for a 12oz serving size.

The dataset for this notebook was obtained at https://www.kaggle.com/starbucks/starbucks-menu/.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

First we read the csv file into a pandas DataFrame to see the data we are working with in an organized way. 

```python
df = pd.read_csv("starbucks_drinkMenu_expanded.csv")
```

```python
df.dtypes
```

```python
df.columns
```

A few of the data types are incorrect as only the first 3 columns should be objects and the other columsn should be integers or floats.

```python
df[' Total Fat (g)'].unique()
```

```python
df[' Total Fat (g)'].replace({'3 2': '3.2'}, inplace = True)
df[' Total Fat (g)'] = df[' Total Fat (g)'].astype(float)
df
```

There was an error in the recording of the data with the most likely excuse is a missing decimal point. The following 4 columns were changed to a float and changed to decimal points instead to work with the other columns.

```python
df['Vitamin A (% DV) '] = df['Vitamin A (% DV) '].str.rstrip('%').astype('float') / 100.0
df['Vitamin C (% DV)'] = df['Vitamin C (% DV)'].str.rstrip('%').astype('float') / 100.0
df[' Calcium (% DV) '] = df[' Calcium (% DV) '].str.rstrip('%').astype('float') / 100.0
df['Iron (% DV) '] = df['Iron (% DV) '].str.rstrip('%').astype('float') / 100.0
```

```python
df['Caffeine (mg)'].unique()
```

Some entries in the Caffeine column had strings instead of a value. To change this, the values were changed to empty and filled in with the median values.

```python
df['Caffeine (mg)'].replace({'varies': np.nan , 'Varies': np.nan}, inplace = True)
```

```python
df['Caffeine (mg)'].fillna(df['Caffeine (mg)'].median(), inplace = True)
```

```python
df['Caffeine (mg)'] = df['Caffeine (mg)'].astype(float)
```

```python
df.dtypes
```

```python
bins = 10
cal = plt.hist(df.Calories, bins = bins)
plt.xlabel('Calories')
plt.ylabel('Number of Drinks')
plt.title('Number of Drinks with certain calories')
for i in range(bins):
    plt.text(cal[1][i],cal[0][i],str(cal[0][i]))
```

16/242 drinks will have 50 Calories or less

15/242 drinks will have more than 350 Calories

Most drinks will be between 100 and 200 Calories

```python
plt.figure(figsize=(14,8))
sns.heatmap(df.corr(),vmin=0, vmax=1, cmap='coolwarm', annot=True)
```

The goal of Clustered Heatmap is to build associations between both the data points and their features. This type of heatmap implements clustering as part of the process of grouping similar features. Most standouts of the heatmap are the association with calories and cholestorol, calories and sugars, sugars with cholestorol. Most of these are obvious with basic nutrition knowledge.

```python
sns.pairplot(df)
```

A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.

```python
grid = sns.JointGrid(x = df.Calories, y = df[' Sugars (g)'], height=8, ratio=50)
grid.plot_joint(plt.scatter)
```

Now we generate a scatter plot similar to the one created in the pairplot but bigger for visualization.  We see that sugar and calories are nearly proportionally related.
