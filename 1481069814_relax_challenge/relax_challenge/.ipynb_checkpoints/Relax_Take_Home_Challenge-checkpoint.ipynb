```python
#rename column for matching names
df_user = df_user.rename({'object_id':'user_id'})
```

```python
#change column to datetime
df_usereng['time_stamp'] = pd.to_datetime(df_usereng['time_stamp'])
```

```python
#group by user_id
grouped_engagement = df_usereng.groupby(['user_id','time_stamp']).sum()
grouped_engagement
```

```python
level_values = grouped_engagement.index.get_level_values
grouped_engagement = grouped_engagement.groupby([level_values(0)] + [pd.Grouper(freq='W', level=-1)]).sum()
```

```python
grouped_engagement.head(20)
```

```python
max_weekly_visits = grouped_engagement.groupby('user_id').max().reset_index()
```

```python
merged = df_user.merge(max_weekly_visits, on='user_id', how='inner')
```
