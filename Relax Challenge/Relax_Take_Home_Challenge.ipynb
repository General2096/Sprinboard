```python
import pandas as pd
import numpy as np
```

```python
df_usereng = pd.read_csv('takehome_user_engagement.csv')
df_user = pd.read_csv('takehome_users.csv',encoding='latin-1')
```

```python
#rename column for matching names
df_user = df_user.rename(columns = {'object_id':'user_id'})
```

```python
#change column to datetime
df_usereng['time_stamp'] = pd.to_datetime(df_usereng['time_stamp'])
```

```python
#group by user_id and timestamp
group_eng = df_usereng.groupby(['user_id','time_stamp']).sum()
group_eng.head(20)
```

```python
#level_values(0) 
#is the user_id, each repeated based on how many sign-ins
level_values = group_eng.index.get_level_values
level_values
```

```python
#this will condense the list based on the number of visits for a 
#week period with the final date being showcased with the total number
#of visits in that week

#pd.Grouper(freq='W', level=-1)
#groupby instruction for an object, groups per week on the last column
group_eng = group_eng.groupby([level_values(0)] + [pd.Grouper(freq='W', level=-1)]).sum()
group_eng.head(20)
```

```python
#condense the list to just show the largest number of visits per user
max_weekly_visits = group_eng.groupby('user_id').max()
max_weekly_visits
```

```python
#combine dataframes on user_id
merged = df_user.merge(max_weekly_visits, on='user_id', how='inner')
```

```python
merged.head()
```

```python
merged.isna().sum()
```

```python
#fill missing values with zero
merged['invited_by_user_id'] = merged.invited_by_user_id.fillna(0)
```

```python
#look at the domain of the email
merged['email_domain'] = merged['email'].apply(lambda x: x.split('@')[-1])
```

```python
#created column adopted based on the number of visits
merged['adopted'] = merged.visited.apply(lambda x: int(x >= 3))
```

```python
merged.head()
```

```python
#columns to not seem to showcase any useful information so they will 
#be removed
merged.drop(columns=['name', 'email', 'visited'],inplace=True)
```

```python
#create dummy columns for its possible options
merged = pd.get_dummies(merged, columns=['creation_source', 'email_domain'])
```

```python
merged.head()
```

```python
#convert both columns to datetime format
merged['creation_time'] = pd.to_datetime(merged['creation_time'])
merged['last_session_creation_time'] = pd.to_datetime(merged['last_session_creation_time'],unit='s')
```

```python
last_ts = merged.last_session_creation_time.max()
last_ts
```

```python
#everything needs to be in a usable format for the random forest classifier
#take the fist 'last_ts' value and subtract the creation time
#and convert it to seconds for each cell in the column
#have everything still in the same relation to one another

#change creation time to the difference between it and the latest session 
#creation
merged['creation_time'] = merged['creation_time'].apply(lambda x: (last_ts - x).total_seconds())
#change last_session to the difference between it and the latest session
#creation
merged['last_session_creation_time'] = merged['last_session_creation_time'].apply(lambda x: (last_ts - x).total_seconds())
```

```python
for column_name in merged.columns:
    column = merged[column_name]
    # Get the count of Zeros in column 
    count = (column == 0).sum()
    # Total number of rows is 8823
    # Anything more than 8817 zeros will have the column dropped
    if abs(count - 8822) < 5:
        merged.drop(columns = [column_name], inplace = True)
    print('Count of zeros in column ', column_name, ' is : ', count)
    continue
```

```python
#Simple classifier
from sklearn.ensemble import RandomForestClassifier

Y = merged['adopted']
X = merged[[x for x in merged.columns if x != 'adopted']]
```

```python
clf = RandomForestClassifier()
clf.fit(X,Y)
```
